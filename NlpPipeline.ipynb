```python
import pandas as pd

# Login using e.g. `huggingface-cli login` to access this dataset
splits = {
    'train': 'data/train-00000-of-00001.parquet',
    'eval': 'data/eval-00000-of-00001.parquet',
    'test': 'data/test-00000-of-00001.parquet'
}

# df = pd.read_parquet("hf://datasets/vic35get/nhtsa_complaints_dataset/" + splits["train"])

df = pd.concat(
    [
        pd.read_parquet(f"hf://datasets/vic35get/nhtsa_complaints_dataset/{splits[s]}")
        for s in ["train", "eval", "test"]
    ],
    ignore_index=True,
)
```

```python
df.head()
```

```python
df = df[['summary', 'label']]
df.head()
```

```python
df['summary'].iloc[1]
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
import re


def clean_text(text):
    text = str(text).lower()
    text = text.replace('tl*', '')
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
df['summary'] = df['summary'].apply(clean_text)
df.head()
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    df['summary'], df['label'], test_size=0.2
)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
clf = RandomForestClassifier(n_estimators=10)
clf.fit(X_train_vec, y_train)
```

```python
pred = clf.predict(X_test_vec)
```

```python
print(classification_report(y_test, pred))
```

```python
def predict(user_desc):
    text = str(user_desc).lower()
    text = text.replace('tl*', '')
    text = re.sub(r'\s+', ' ', text).strip()

    text_vec = vectorizer.transform([text])
    pred = clf.predict(text_vec)[0]
    return pred
```

```python
print(predict(''))
```

```python
print(predict('my car'))
```

```python
print(predict('my car is dead'))
```

```python
print(predict('the wheel fell off'))
```
